Here I'm going to convert the output of `rake to_json` to a sparse matrix containing feature "cluster" IDs as columns, genomes as rows and cells are the # of times a cluster appears in a genome. This way I can use some sort of data mining techniques in Python or R.

```python
import json
from collections import defaultdict
import pandas as pd
import cPickle
```

```python
genomes = []
with open('../db.json') as handle:
    for line in handle:
        genome = json.loads(line)
        genomes.append(genome)
```

```python
for genome in genomes:
    print genome['organism']
```

```python
# I want to see how connected the graph is.
# Do protein relationships form discrete or fuzzy, overlapping clusters?

clusters = set()

for genome in genomes:
    for feature in genome['features']:
        cluster_members = tuple( i['id'] for i in feature['related'] )
        clusters.add(cluster_members)
```

```python
# give each cluster an id
cluster_to_id = {}

# members -> cluster_id
for i, k in enumerate(clusters):
    cluster_to_id[k] = i
    
# feature -> [ cluster ids ]
p_to_clusters = defaultdict(set)
for m in clusters:
    for p in m:
        p_to_clusters[p].add(cluster_to_id[m])
```

```python
genome_cluster_count = defaultdict(lambda: defaultdict(lambda: 0))
for genome in genomes:
    for feature in genome['features']:
        clusters = p_to_clusters[feature['id']]
        for cluster in clusters:
            genome_cluster_count[genome['id']][cluster] += 1
```

```python
df = pd.DataFrame.from_dict(genome_cluster_count).T.fillna(0)
```

```python
# and now we have our big, sparse-matrix!
df.index = [ g['organism'] for g in genomes ]
df
```

```python
# save it to a pickle.

with open('genomes-cluster-matrix.pickle', 'w') as handle:
    cPickle.dump(df, handle)
    
# save it to CSV

df.to_csv('genomes_cluster_matrix.csv')
```
