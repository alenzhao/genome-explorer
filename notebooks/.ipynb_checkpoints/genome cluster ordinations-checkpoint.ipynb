```python
import cPickle
import pandas as pd

from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.manifold import TSNE, Isomap, MDS
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder

import pylab

import matplotlib.pyplot as plt

# use seaborn plotting defaults
# If this causes an error, you can comment it out.
import seaborn as sns
sns.set()
```

```python
# load data

df = pd.read_csv('genome-cluster-table.csv', index_col=0)
genus = set( i.split()[0] for i in df.index )
print genus

roides = [ i for i in df.index if 'Bacteroides' in i ]
df = df.ix[roides]
species = [ i.split()[1] for i in df.index ]
```

```python
fr = TruncatedSVD(n_components=10)
fitted = fr.fit_transform(df.values)
```

```python
plt.plot(fr.explained_variance_ratio_)
```

```python
lcol = LabelEncoder().fit_transform(species)
plt.scatter(fitted[:,1], fitted[:,2], c = lcol, cmap=pylab.cm.cool)
plt.axis('equal')
plt.show()
print lcol
```
