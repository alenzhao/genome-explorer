# generate cluster contigency table from uclust output

before running do:
    
```sh
rake dump:proteins > proteins.fasta
rake dump:genomes > genomes.csv
rake dump:features > features.csv

sh cluster_usearch.sh
```

Then run this notebook to generate the table as `genome-cluster-table.csv`.

Cluster #s are assigned arbitrarily and are subject to change every time you run `cluster_usearch.sh`

````python
import pandas as pd
````

````python
features = pd.DataFrame.from_csv('../features.csv')
genomes = pd.DataFrame.from_csv('../genomes.csv')

clusters = pd.read_csv('../clusters.0.2.uc', sep="\t", comment="#", header=None)

clusters = clusters[[0, 1, 8]]

clusters.columns = ['cluster_type', 'cluster', 'feature']
clusters = clusters[clusters['cluster_type'].isin(['H', 'S'])]
clusters = clusters.set_index('feature')

proteins = features.ix[features.type == 'CDS']
````

````python
mapping = proteins.join(clusters)[['genome_id', 'cluster']]

# group by genome and cluster, count number of features
dat = mapping.groupby(['genome_id', 'cluster']).agg(len).unstack().fillna(0)
````

````python
# write to table
dat.to_csv('genome-cluster-table.csv')
````

````python
# clusters per genome
dat.sum(axis=1).hist()
````

````python
# cluster size
dat.sum(axis=0).hist()
````

````python
# "core" genome
core = dat[dat.columns[dat.min(axis=0) > 0]]
````
